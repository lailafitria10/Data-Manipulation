Anda adalah seorang dosen di kelas dasar pemrograman
Ada 3 nilai berikut : 
np.random.normal(mean, var, n)
- tugas ~ N(85, 7)
- uts ~ N(60, 15)
- uas ~ N(75, 10)

buat kolom baru (nilai akhir) yang dihitung dengan formula :
tugas : 20%
UTS : 40%
UAS : 40%

buat kolom baru dan lakukan binning dengan kriteria berikut:
A ~ 85-100
A- ~ 80-85
B+ ~ 75-80
B ~ 70-75
B- ~ 65-70
C+ ~ 60-65
C ~ 55-60
C- ~ 50-55
D ~ 20-50
E ~ 0-20

```python
tugas = np.random.normal(85, 7, 40)
uts = np.random.normal(60, 15, 40)
uas = np.random.normal(75, 10, 40)
```

```python
df_nilai = pd.DataFrame({'tugas':tugas,
                         'uts':uts,
                         'uas':uas})
```

```python
df_nilai['nilai_akhir'] = 0.2*df_nilai['tugas'] + 0.4*df_nilai['uts'] + 0.4*df_nilai['uas']
```

```python
bins = [0,20,50,55,60,65,70,75,80,85,100]
labels = ['E', 'D', 'C-', 'C', 'C+', 'B-', 'B', 'B+', 'A-', 'A']
df_nilai['nilai_huruf'] = pd.cut(df_nilai['nilai_akhir'], bins, labels=labels)
```

```python
df_nilai[df_nilai['nilai_huruf'] == 'A']
```
